```python
# Create a new column based on conditions
ad_data['ROAS'] = ad_data['Revenue (USD)'] / ad_data['Spend (USD)']
ad_data['High Performer'] = ad_data['ROAS'] > 2  # Boolean column
```

```python
ad_data.head(1)
```

```python
# Apply a custom function to a column
def categorize_revenue(x):
    if x > 1000:
        return 'High'
    else:
        return 'Low'

ad_data['Revenue Category'] = ad_data['Revenue (USD)'].apply(categorize_revenue)
```

```python
linkedin_variant = ad_data[(ad_data['Ad Platform'] == 'LinkedIn') & (ad_data['Campaign Name'] == 'B2B Outreach') ]
```

```python
linkedin_variant.head(10)
```

```python
# Full outer merge on 'Campaign Name'
comparison_df = linkedin.merge(linkedin_variant, on='Campaign Name', how='outer', suffixes=('_df1', '_df2'), indicator=True)
```

```python
# Check for exact matches (all columns)
matching_rows = comparison_df[
    (comparison_df['Region_df1'] == comparison_df['Region_df2']) &
    (comparison_df['Spend (USD)_df1'] == comparison_df['Spend (USD)_df2']) &
    (comparison_df['Revenue (USD)_df1'] == comparison_df['Revenue (USD)_df2'])
]

```

```python
matching_rows
```

```python
# Check for non-matching rows
non_matching_rows = comparison_df[
    ~(
        (comparison_df['Ad Platform_df1'] == comparison_df['Ad Platform_df2']) &
        (comparison_df['Spend (USD)_df1'] == comparison_df['Spend (USD)_df2']) &
        (comparison_df['Revenue (USD)_df1'] == comparison_df['Revenue (USD)_df2'])
    )
]
```
